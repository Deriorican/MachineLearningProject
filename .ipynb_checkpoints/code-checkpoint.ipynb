Get rid of useless/unusable features and re-working other features to make them usable

```python
X1.drop(["img_url", "description", "title", "is_adult"], axis=1, inplace=True)
X2.drop(["img_url", "description", "title", 'is_adult'], axis=1, inplace=True)
```

```python
# Replacing the "genres" feature : Instead of a string containing the genres of the movie,
# we replace it by an array of ones and zeros. Ones are put on the indexes of the corresponding 
# genre of the movies.

index = 0
genresDic = {}
for items in X1.get("genres"):
    items = items.split(",")
    for item in items:
        if item not in genresDic and item != "Unknown":
            genresDic[item] = index
            index += 1
newGenres = []
for items in X1.get("genres"):
    items = items.split(",")
    newGenres.append(np.zeros(len(genresDic)))
    for item in items:
        if item != "Unknown":
            newGenres[-1][genresDic[item]] = 1

# Replacing the "studio" feature : Instead of a string containing the studio of the movie,
# we replace it by an array of ones and zeros. Ones are put on the indexes of the corresponding 
# studio of the movies.
        
index = 0
studDic = {}
for item in X1.get("studio"):
    if item not in studDic:
        studDic[item] = index
        index += 1
newStudios = []
for item in X1.get("studio"):
    newStudios.append(np.zeros(len(studDic)))
    newStudios[-1][studDic[item]] = 1

X1["genres"] = newGenres
X1["studio"] = newStudios
X1
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing


newFeatures = []

feature = "img_embeddings"
pcaX1 = np.zeros((len(X1[feature]), len(np.array(X1[feature][0][1:-1].split(",")).astype(float))))
for i, j in zip(X1[feature], range(len(X1[feature]))):
    pcaX1[j] = np.array(i[1:-1].split(",")).astype(float)

pcaX1 = preprocessing.scale(pcaX1)
pca = PCA(n_components=100)
pca.fit(pcaX1)
newFeatures.append(pca.transform(pcaX1))

feature = "text_embeddings"
pcaX1 = np.zeros((len(X1[feature]), len(np.array(X1[feature][0][1:-1].split(",")).astype(float))))
for i, j in zip(X1[feature], range(len(X1[feature]))):
    pcaX1[j] = np.array(i[1:-1].split(",")).astype(float)

pcaX1 = preprocessing.scale(pcaX1)
pca = PCA(n_components=9)
pca.fit(pcaX1)
newFeatures.append(pca.transform(pcaX1))

newFeatures.append(np.vstack(np.array(X1["genres"])))
newFeatures.append(np.vstack(np.array(X1["studio"])))
```

```python

features = ["img_embeddings", "text_embeddings", "genres", "studio"]
for newFeature, feature in zip(newFeatures, features):
    for i in range(len(newFeature[0])):
        X1[feature + str(i)] = newFeature[:, i]
    X1.drop(feature, axis=1, inplace=True)
```

```python
excluded_features = ['is_adult', 'studio', 'genre']
for feature in X1.keys():
    excluded = False
    for excluded_feature in excluded_features :
        if excluded_feature in feature:
            excluded = True
    if not excluded :
        X1[feature] = (X1[feature] - X1[feature].mean()) / X1[feature].std()
X1

Y1 = (Y1- Y1.mean()) / Y1.std()
```

```python
corr = X1.corr()
threshold = 0.8

feature_to_delete = []
print(type(corr))
corr_max = 0
for feature in X1.keys():
    for feature2 in X1.keys():
        if feature == feature2 :
            break
            
        if abs(corr[feature][feature2]) > threshold :
            feature_to_delete.append((feature, feature2))
        if abs(corr[feature][feature2]) > corr_max:
            corr_max = abs(corr[feature][feature2])
            
print(feature_to_delete)
print(corr_max)
        
#sns.heatmap(abs(corr))

#print(X1['ratings'].to_frame())
corr = X1['ratings'].to_frame().corrwith(Y1)
sorted_arg = np.flip(np.array(np.argsort(corr)))
# print(sorted_arg)
print(corr)
# print(Y1)
```
